```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

def make_dirs(path):
    try:
        if not os.path.exists(path):
            os.makedirs(path)
            print(f'{path}建立成功.')
        else:
            print(f'{path}已存在.')
    except Exception as e:
        print('建立floder失敗.', e)

def get_soup(url, post_data=None):
    try:
        if post_data is not None:
            resp = requests.post(url, post_data, headers={
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.2'})
        else:
            resp = requests.get(url, headers={
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.2'})
        resp.encoding = 'utf-8'
        if resp.status_code == 200:
            soup = BeautifulSoup(resp.text, 'lxml')
            return soup
        else:
            print('網頁取得失敗', resp.status_code)
    except Exception as e:
        print('網址錯誤', e)

def save_pic(url, filename='temp.jpg'):
    try:
        resp = requests.get(url, headers={
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.2'})
        if resp.status_code == 200:
            with open(filename, 'wb') as f:
                f.write(resp.content)
                print(f'{filename}儲存完畢')
    except Exception as e:
        print('儲存失敗', resp.status_code, e)
```

```python
data = []

pg = chrome.page_source
soup = BeautifulSoup(pg, 'lxml')
bb = soup.find("div", class_="searchResultsContent").find_all("div", class_="col_academicPresent")

for article in bb:
    h3 = article.find("h3")
    try:
        # 檢查是否找到 h3 元素
        if h3:
            input_text = str(h3)

            # 使用正規表達式進行匹配
            match = re.search(r"'(.*?)'", input_text)

            if match:
                result = match.group(1)
                print(result)

                # 將提取的部分添加到 URL 字符串中
                url = 'https://www.airitilibrary.com/Article/Detail?DocID=' + result
                print(url)

                # 使用 get_soup 函數抓取網頁
                resp = get_soup(url)

                # 在抓取的網頁中找到相關內容
                if resp:
                    abstract_element = resp.find("div", class_="academicAbstractGroup")
                    if abstract_element:
                        abstract = abstract_element.find("p").text
                        print(abstract)
                        
                        # 同時抓取標題和作者，並添加到 data 中
                        title = article.find("h3").text.strip()
                        author = article.find("span", class_="author").text.strip()
                        data.append([title, author, url, abstract])
                    else:
                        print("未找到摘要元素")
                        # 若沒有摘要，將 abstract 設為 None
                        title = article.find("h3").text.strip()
                        author = article.find("span", class_="author").text.strip()
                        data.append([title, author, url, None])
                else:
                    print("網頁抓取失敗")
            else:
                print("未找到匹配的內容")
        else:
            print("未找到 h3 元素")
    except:
        print(None)

# 打印整理後的 data
print(data)
```
